```python
from pomegranate import BayesianNetwork, DiscreteDistribution, State

# Define the Bayesian network
bn = BayesianNetwork("Software Testing")

# Define the discrete distributions for each variable
A = DiscreteDistribution({"pass": 0.9, "fail": 0.1})
B = DiscreteDistribution({"pass": 0.95, "fail": 0.05})
C = DiscreteDistribution({"pass": 0.8, "fail": 0.2})
D = DiscreteDistribution({"pass": 0.9, "fail": 0.1})
E = DiscreteDistribution({"pass": 0.95, "fail": 0.05})
F = DiscreteDistribution({"pass": 0.8, "fail": 0.2})
G = DiscreteDistribution({"pass": 0.9, "fail": 0.1})
H = DiscreteDistribution({"pass": 0.95, "fail": 0.05})
I = DiscreteDistribution({"pass": 0.8, "fail": 0.2})

# Create the states for each variable
state_a = State(A, name="A")
state_b = State(B, name="B")
state_c = State(C, name="C")
state_d = State(D, name="D")
state_e = State(E, name="E")
state_f = State(F, name="F")
state_g = State(G, name="G")
state_h = State(H, name="H")
state_i = State(I, name="I")

# Add the states to the network
bn.add_states(state_a, state_b, state_c, state_d, state_e, state_f, state_g, state_h, state_i)

# Add the edges between the variables
bn.add_edge(state_a, state_d)
bn.add_edge(state_b, state_d)
bn.add_edge(state_c, state_e)
bn.add_edge(state_d, state_f)
bn.add_edge(state_e, state_f)
bn.add_edge(state_d, state_g)
bn.add_edge(state_f, state_g)
bn.add_edge(state_g, state_i)
bn.add_edge(state_h, state_i)

# Finalize the structure of the network
bn.bake()

# Define the evidence for the network
evidence = {
    "A": "pass",
    "B": "fail",
    "C": "pass",
    "H": "fail"
}
# Define the evidence for the network for different test cases
test_case1 = {
    "A": "pass",
    "B": "fail",
    "C": "pass",
    "H": "fail"
}

test_case2 = {
    "A": "pass",
    "B": "fail",
    "C": "pass",
    "H": "pass"
}

test_case3 = {
    "A": "fail",
    "B": "fail",
    "C": "fail",
    "H": "fail"
}

# Compute the probabilities for the query variable for each test case
query1 = bn.predict_proba(test_case1)
query2 = bn.predict_proba(test_case2)
query3 = bn.predict_proba(test_case3)

# Print the results for each test case
print("Test Case 1:")
print("Probability of test case passing:", query1[8].parameters[0]["pass"])
print("Probability of test case failing:", query1[8].parameters[0]["fail"])

print("Test Case 2:")
print("Probability of test case passing:", query2[8].parameters[0]["pass"])
print("Probability of test case failing:", query2[8].parameters[0]["fail"])

print("Test Case 3:")
print("Probability of test case passing:", query3[8].parameters[0]["pass"])
print("Probability of test case failing:", query3[8].parameters[0]["fail"])


# Compute the probabilities for the query variable
query = bn.predict_proba(evidence)

# Print the results
print("Test Case 4:")
print("Probability of test case passing:", query[8].parameters[0]["pass"])
print("Probability of test case failing:", query[8].parameters[0]["fail"])
```
